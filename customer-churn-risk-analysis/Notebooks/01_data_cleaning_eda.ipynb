```python
import pandas as pd

# Load the dataset
# Example of an absolute path
df = pd.read_csv(r'C:\Users\Gifty\Projects\Customer Segmentation\data\raw\OnlineRetail.csv', encoding='ISO-8859-1')

# Display the first few rows
df.head()
```

```python
# Dataset is loaded
```

```python
# Checking for the null values
df.isnull().sum()
```

```python
# Drop missing customers
df = df.dropna(subset=['CustomerID'])
```

```python
# Checking for the null values
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Convert date
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.dtypes
```

```python
# Remove invalid prices
df = df[df['UnitPrice'] > 0]

# Total price
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
from pathlib import Path

# Create a Path object
path = Path("data/processed/cleaned_transactions.csv")

# Create the parent directory (parents=True creates data AND processed if both are missing)
path.parent.mkdir(parents=True, exist_ok=True)

# Save
df.to_csv(path, index=False)
```
