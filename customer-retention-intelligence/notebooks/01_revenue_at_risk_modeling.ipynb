```python
import pandas as pd

churn = pd.read_csv("C:\\Users\\GIFTY\\Projects\\customer-analytics-portfolio\\customer-churn-risk-analysis\\data\\processed\\churn_scoring_dataset.csv")
clv = pd.read_csv("C:\\Users\\GIFTY\\Projects\\customer-analytics-portfolio\\customer-lifetime-value-analysis\\data\\processed\\clv_scoring_dataset.csv")

final_df = churn.merge(clv, on="CustomerID", how="inner")
```

```python
final_df["revenue_at_risk"] = (
    final_df["churn_probability"] * final_df["clv_6m"]
)
```

```python
final_df["retained_value"] = (
    (1 - final_df["churn_probability"]) * final_df["clv_6m"]
)
```

```python
final_df.to_csv("C:\\Users\\GIFTY\\Projects\\customer-analytics-portfolio\\customer-retention-intelligence\\data\\processed\\integrated_retention_dataset.csv", index=False)
```
